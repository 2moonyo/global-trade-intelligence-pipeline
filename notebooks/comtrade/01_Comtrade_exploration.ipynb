```python
import pandas as pd
import json
from pathlib import Path
from typing import List, Dict, Any

def load_comtrade_data_to_dataframe(base_path: str) -> pd.DataFrame:
    """
    Recursively reads all JSON files from the directory structure
    and combines them into a single pandas DataFrame.
    
    Args:
        base_path: Root directory containing the JSON files
        
    Returns:
        pandas DataFrame with all records from all JSON files
    """
    # Find all JSON files recursively
    json_files = list(Path(base_path).rglob("*.json"))
    print(f"Found {len(json_files)} JSON files")
    
    all_records = []
    
    # Read each JSON file
    for idx, json_file in enumerate(json_files, 1):
        try:
            with open(json_file, 'r') as f:
                data = json.load(f)
            
            # Extract the 'data' array from the JSON
            if isinstance(data, dict) and 'data' in data:
                records = data['data']
                
                # Get the count from the JSON file
                total_count = data.get('count', None)
                
                # Optionally add metadata from the file
                # (filename, year from path structure, etc.)
                for record in records:
                    record['source_file'] = json_file.name
                    record['source_year'] = json_file.parent.name
                    record['total_count'] = total_count
                
                all_records.extend(records)
                
                if idx % 10 == 0:  # Progress indicator
                    print(f"Processed {idx}/{len(json_files)} files, "
                          f"total records: {len(all_records)}")
            else:
                print(f"Unexpected structure in {json_file}")
                
        except Exception as e:
            print(f"Error reading {json_file}: {e}")
            continue
    
    # Convert to DataFrame
    print(f"\nCreating DataFrame with {len(all_records)} total records...")
    df = pd.DataFrame(all_records)
    
    print(f"DataFrame shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    
    return df

# Usage
base_path = "/Users/chromazone/Documents/Python/Data Enginering Zoomcamp/Capstone_monthly/data/bronze/comtrade/monthly_history"
df = load_comtrade_data_to_dataframe(base_path)

# Display info
print("\nDataFrame Info:")
print(df.info())
print("\nFirst few rows:")
print(df.head())
print("\nMemory usage:")
print(f"{df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```
